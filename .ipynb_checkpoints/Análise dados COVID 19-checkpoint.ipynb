#### Importar Dados de uma API com informações dos casos de COVID no Brasil
- Analisar dados fornecidos por uma API, para visualizar gráficos através do quickchart comparando casos Confirmados vs Recuperados e Óbitos vs Recuperados
- Gerar Qr Code para visualizar o gráfico

```python
import requests as r
```

```python
url = 'https://api.covid19api.com/dayone/country/brazil'
resp = r.get(url)
```

```python
resp.status_code
```

```python
raw_data = resp.json()
```

```python
raw_data[0]
```

#### Selecionado apenas as colunas que serão utilizadas na análise

```python
final_data = []
for obs in raw_data:
    final_data.append([obs['Confirmed'], obs['Deaths'], obs['Recovered'], obs['Active'], obs['Date']])
```

#### Adicionando o header 

```python
final_data.insert(0, ['Confirmados', 'Obitos', 'Recuperados', 'Ativos', 'Data'])
final_data
```

```python
CONFIRMADOS = 0
OBITOS = 1
RECUPERADOS = 2
ATIVOS = 3
DATA = 4
```

#### Selecionando apenas os elementos que correspondem ao dia, mês e ano da coluna data

```python
for i in range(1, len(final_data)):
    final_data[i][DATA] = final_data[i][DATA][:10]
```

```python
final_data
```

#### Salvando os dados em um csv

```python
import csv
```

```python
with open('brasil-covid.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerows(final_data)
```

#### Convertendo para datetime

```python
import datetime as dt
```

```python
for i in range(1, len(final_data)):
    final_data[i][DATA] = dt.datetime.strptime(final_data[i][DATA], '%Y-%m-%d')
```

```python
final_data
```

#### Renderizar gráfico utilizando quickchart

- Função que constrói contendo dados reais de y e os labels de cada um dos dados 

```python
def get_datasets(y, labels):
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                'label':labels[i],
                'data': y[i]
            })
        return datasets
    else:
        return [
            {
                'label': labels[0],
                'data': y
            }
        ]
```

- Função para definir o título do gráfico

```python
def set_title(title=''):
    if title != '':
        display = 'true'
    else:
        display = 'false'
    return{
        'title':title,
        'display': display
    }
```

- Função que cria o dicionário que representa o gráfico

```python
def create_chart(x, y, labels, kind='bar', title=''):
    datasets = get_datasets(y, labels)
    options = set_title(title)
    
    chart = {
        'type': kind,
        'data':{
            'labels': x,
            'datasets': datasets
        },
        'options': options
    }
    return chart
```

- Função que realiza a requisição na API

```python
def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    resp = r.get(f'{url_base}?c={str(chart)}')
    return resp.content
```

- Função para salvar a imagem retornada da api

```python
def save_image(path, content):
    with open (path, 'wb') as image:
        image.write(content)
```

- Função para mostrar a imagem salva do retorno da api

```python
from PIL import Image
from IPython.display import display
```

```python
def display_image(path):
    img_pil = Image.open(path)
    display(img_pil)
```

#### Plote de barras em grupos (a cada 20 dias) mostrando a evolução de casos confirmados e recuperados

```python
y_data_1 = []
for obs in final_data[1::20]:
    y_data_1.append(obs[CONFIRMADOS])
    
y_data_2 = []
for obs in final_data[1::20]:
    y_data_2.append(obs[RECUPERADOS])
    
labels = ['Confirmados', 'Recuperados']

x = []
for obs in final_data[1::20]:
    x.append(obs[DATA].strftime('%d/%m/%Y'))
    
chart = create_chart(x, [y_data_1, y_data_2], labels, title='Gráfico Confirmados vs Recuperados')
chart_content = get_api_chart(chart)
save_image('grafico.png', chart_content)
display_image('grafico.png')
```

- Função para gerar Qr Code

```python
from urllib.parse import quote
```

```python
def get_api_qrcode(link):
    text = quote(link)
    url_base = 'https://quickchart.io/qr'
    resp = r.get(f'{url_base}?text={text}')
    return resp.content
```

```python
url_base = 'https://quickchart.io/chart'
link = f'{url_base}?c={str(chart)}'
save_image('qr-code.png', get_api_qrcode(link))
display_image('qr-code.png')
```

#### Plote de barras em grupos (a cada 20 dias) mostrando a evolução de óbitos vs casos recuperados

```python
y_data_3 = []
for obs in final_data[1::20]:
    y_data_3.append(obs[OBITOS])
    
y_data_4 = []
for obs in final_data[1::20]:
    y_data_4.append(obs[RECUPERADOS])
    
labels = ['Óbitos', 'Recuperados']

x = []
for obs in final_data[1::20]:
    x.append(obs[DATA].strftime('%d/%m/%Y'))
    
chart = create_chart(x, [y_data_3, y_data_4], labels, title='Gráfico Óbitos vs Recuperados')
chart_content = get_api_chart(chart)
save_image('grafico2.png', chart_content)
display_image('grafico2.png')
```
